## Resignation trends in Department of Education, Training and Development (DETE) and Technical and Further Education (TAFE) 

The following questions will be explored by looking at exit survey results from both departments.

1) Are employees who only worked for the institutes for a short period of time resigning due to some kind of dissatisfaction? 

2) What about employees who have been there longer?

To answer these questions, as mentioned earlier, both datasets from DETE and TAFE will be compared together.

TAFE exit survey obtained from: https://data.gov.au/dataset/ds-qld-89970a3b-182b-41ea-aea2-6f9f17b5907e/details?q=exit%20survey

DETE exit survey obtained from: https://data.gov.au/dataset/ds-qld-fe96ff30-d157-4a81-851d-215f2a0fe26d/details?q=exit%20survey

### 2) Determining the important columns for this analysis:


Since we only need to find out the reason for leaving, and information about the individual(i.e ID, Institute, Workarea, time of employment, gender etc), we do not need columns pertaining to the institute views of these individuals. So we will drop these columns for both datasets.

### 3) Renaming Columns:

For columns in dete_survey, making all columns lowercase, removing whitespace from end of strings and replacing spaces with underscores.

```python
dete_survey_updated.columns = dete_survey_updated.columns.str.replace('.','').str.replace('-','').str.strip()
dete_survey_updated.columns = dete_survey_updated.columns.str.replace(' ', '_').str.lower()
dete_survey_updated = dete_survey_updated.rename(columns={'separationtype':'separation_type','health_&_safety':'health&safety'})
print(dete_survey_updated.columns)
```

```python
print(dete_survey_updated.isnull().sum())
```

```python
print(tafe_survey_updated.isnull().sum())
```

We want to rename the columns in TAFE survey such that the similar columns with DETE survey have the same name. Non-similar columns are given neater names(e.g. length of service columns)

```python
tafe_survey_updated.columns = tafe_survey_updated.columns.str.strip()
tafe_columns ={'Record ID':'id',
'Institute': 'institute',
'WorkArea':'business_unit',
'CESSATION YEAR':'cease_date',
'Reason for ceasing employment':'seperation_type',
'Contributing Factors. Career Move - Public Sector':'career_move_to_public_sector',
'Contributing Factors. Career Move - Private Sector':'career_move_to_private_sector',
'Contributing Factors. Career Move - Self-employment':'career_move_to_self_employment',
'Contributing Factors. Ill Health':'ill_health',
'Contributing Factors. Maternity/Family':'maternity/family',
'Contributing Factors. Dissatisfaction':'dissatisfaction_with_the_department',
'Contributing Factors. Job Dissatisfaction':'job_dissatisfaction',
'Contributing Factors. Interpersonal Conflict':'interpersonal_conflicts',
'Contributing Factors. Study':'study',
'Contributing Factors. Travel':'travel',
'Contributing Factors. Other':'other',
'Contributing Factors. NONE':'none',
'Gender. What is your Gender?':'gender',
'CurrentAge. Current Age':'age',
'Employment Type. Employment Type':'employment_status',
'Classification. Classification':'classification',
'LengthofServiceOverall. Overall Length of Service at Institute (in years)':'institute_service',
'LengthofServiceCurrent. Length of Service at current workplace (in years)':'role_service'}
tafe_survey_updated = tafe_survey_updated.rename(columns=tafe_columns)

#Combining the "study" and "travel" columns into one column so that it follows the DETE survey's "study/travel" column:
tafe_survey_updated["study/travel"] = tafe_survey_updated["travel"]
tafe_survey_updated.loc[tafe_survey_updated["study"]=="Study", "study/travel"] = "Study"
tafe_survey_updated.drop(['study','travel'], axis=1)

#Combining the "other" and "none" columns into one column so that it matches the DETE survey's "none_of_the_above" column:
tafe_survey_updated["none_of_the_above"] = tafe_survey_updated["none"]
tafe_survey_updated.loc[tafe_survey_updated["other"]=="Other", "none_of_the_above"] = "Other"
tafe_survey_updated.drop(["none","other"], axis=1)

print(tafe_survey_updated.isnull().sum())
```

### 4) Removing rows from dete_start_date, cease_date & role_start_date from the dete_survey_updated dataset that do not make sense:

- Cease date values need to be after start_data values
- Start date values should be > the year 1940 because more people in this field started working in their 20s.

```python
### 1) Dropping NA rows in the cease_date column:
print(dete_survey_updated["cease_date"].value_counts(dropna=False))
dete_survey_updated = dete_survey_updated.loc[dete_survey_updated["cease_date"]!="Not Stated"]
print(dete_survey_updated["cease_date"].value_counts(dropna=False))
```

```python
### 2) Dropping the NA rows in the dete_start_date column:
print(dete_survey_updated["dete_start_date"].value_counts(dropna=False))
dete_survey_updated = dete_survey_updated.loc[dete_survey_updated["dete_start_date"] != "Not Stated"]
dete_survey_updated = dete_survey_updated.loc[dete_survey_updated["role_start_date"] != "Not Stated"]
print(dete_survey_updated["dete_start_date"].value_counts(dropna=False))
print(dete_survey_updated["role_start_date"].value_counts(dropna=False))
```

```python
#3) Changing the cease_date, dete_start_date and role_start_date columns to year in integer form:
dete_survey_updated["cease_date"] = dete_survey_updated["cease_date"].str[-4:]
dete_survey_updated["dete_start_date"] = dete_survey_updated["dete_start_date"]
dete_survey_updated["role_start_date"] = dete_survey_updated["role_start_date"]
dete_survey_updated[["cease_date",'dete_start_date','role_start_date']]=dete_survey_updated[["cease_date",'dete_start_date','role_start_date']].apply(pd.to_numeric)

#4) Checking if cease_date >= dete_start_date:
dete_survey_updated["institute_service"] = dete_survey_updated["cease_date"] - dete_survey_updated["dete_start_date"]
print(dete_survey_updated["institute_service"].value_counts(dropna=False))

#5) Checking if cease_date >= role_start_date:
dete_survey_updated["role_service"] = dete_survey_updated["cease_date"]- dete_survey_updated["role_start_date"]
print(dete_survey_updated["role_service"].value_counts(dropna=False))
dete_survey_updated = dete_survey_updated.loc[dete_survey_updated["role_service"]>=0]
dete_survey_updated = dete_survey_updated.loc[dete_survey_updated["role_service"]<80]
print(dete_survey_updated["role_service"].value_counts(dropna=False))
```

```python
#6) Dropping the role_start_date and dete_start_date columns:
dete_survey_updated = dete_survey_updated.drop(columns = ['role_start_date','dete_start_date'])
dete_survey_updated.head()
```

### 5) Updating the values in the institute_service and role_service columns in the tafe_survey_updated dataset for easy comparison with the dete_survey_updated_dataset:

The values in the above-mentioned columns are in ranges e.g. 7-10.  We need to convert these values into integers for easy comparison with the dete_survey_updated dataset.  So we need to get the average of these numbers:

```python
#1) Splitting the institute service column values by the delimeter '-':
tafe_survey_updated = tafe_survey_updated.loc[tafe_survey_updated["institute_service"].notnull()]
print(tafe_survey_updated["institute_service"].value_counts(dropna=False))
tafe_survey_updated[['first','last']] = tafe_survey_updated["institute_service"].str.split('-', expand=True)
tafe_survey_updated.loc[tafe_survey_updated["first"] =='Less than 1 year','first'] = '0'
tafe_survey_updated.loc[tafe_survey_updated['first']=='More than 20 years','first'] = '21'
tafe_survey_updated[['first','last']] = tafe_survey_updated[['first','last']].apply(pd.to_numeric)
tafe_survey_updated = tafe_survey_updated.drop(columns=['institute_service'])
tafe_survey_updated['institute_service'] = (tafe_survey_updated['first']+tafe_survey_updated['last'])/2
tafe_survey_updated.loc[tafe_survey_updated['institute_service'].isnull(), 'institute_service'] = tafe_survey_updated.loc[tafe_survey_updated['institute_service'].isnull(), 'first']
print(tafe_survey_updated['institute_service'].value_counts(dropna=False))
tafe_survey_updated = tafe_survey_updated.drop(columns = ['first','last'])
```

So far, we have assigned the average of the year range as the value for institute_service.  For the values corresponding to "Less than 1 year", we have modified it to 0 to be consistent with the dete_survey_updated dataset.  Also, for values corresponding to "More than 20 year", we have modified it to 21 years as we cannot ascertain what the value exactly is but it is not important.

```python
#2) Splitting the role_service column with the delimeter, '-':
print(tafe_survey_updated['role_service'].value_counts())
tafe_survey_updated[['first','last']] = tafe_survey_updated["role_service"].str.split('-',expand=True)
tafe_survey_updated.loc[tafe_survey_updated['first']=='Less than 1 year', 'first'] = '0'
tafe_survey_updated.loc[tafe_survey_updated['first']=='More than 20 years', 'first'] = '21'
tafe_survey_updated[['first','last']] = tafe_survey_updated[['first','last']].apply(pd.to_numeric)
tafe_survey_updated = tafe_survey_updated.drop(columns='role_service')
tafe_survey_updated['role_service'] = (tafe_survey_updated['first'] + tafe_survey_updated['last'])/2
tafe_survey_updated.loc[tafe_survey_updated['role_service'].isnull(),'role_service']= tafe_survey_updated.loc[tafe_survey_updated['role_service'].isnull(),'first']
tafe_survey_updated = tafe_survey_updated.drop(columns=['first','last'])
print(tafe_survey_updated['role_service'].value_counts(dropna=False))
```

Similarly, we have assigned the average of the year range as the value for role_service.  For the values corresponding to "Less than 1 year", we have modified it to 0 to be consistent with the dete_survey_updated dataset.  Also, for values corresponding to "More than 20 year", we have modified it to 21 years as we cannot ascertain what the value exactly is but it is not important.

For the Dete survey, the following columns are important to the study:

2) dissatisfaction_with_the_department
3) physical_work_environment
4) lack_of_recognition
5) lack_of_job_security

For the tafe survey:

```python
#Making a new column "dissatisfied" which has a value of True if one of the above-mentioned columns is true for Dete_survey:

list = ['job_dissatisfaction',
'dissatisfaction_with_the_department',
'physical_work_environment',
'lack_of_recognition',
'lack_of_job_security',
'work_location',
'employment_conditions',
'work_life_balance',
'workload']
dete_survey_updated["dissatisfied"] = dete_survey_updated[list].any(axis=1, skipna=False)
dete_survey_updated.head()
```

```python
#Converting values in 'job_dissatisfaction' and 'dissatisfaction_with_the_department' for the tafe_survey_updated dataset into True or False
print(tafe_survey_updated['job_dissatisfaction'].value_counts(dropna=False))
print(tafe_survey_updated['dissatisfaction_with_the_department'].value_counts(dropna=False))
tafe_survey_updated.loc[tafe_survey_updated['dissatisfaction_with_the_department'].str.contains('Contributing Factors', na=False),'dissatisfaction_with_the_department']=True
tafe_survey_updated.loc[tafe_survey_updated['dissatisfaction_with_the_department']=='-','dissatisfaction_with_the_department']=False
tafe_survey_updated.loc[tafe_survey_updated['dissatisfaction_with_the_department'].isnull(),'dissatisfaction_with_the_department']=False
tafe_survey_updated.loc[tafe_survey_updated['job_dissatisfaction']=='Job Dissatisfaction','job_dissatisfaction']=True
tafe_survey_updated.loc[tafe_survey_updated['job_dissatisfaction']=='-','job_dissatisfaction']=False
tafe_survey_updated.loc[tafe_survey_updated['job_dissatisfaction'].isnull(),'job_dissatisfaction']=False
print(tafe_survey_updated['job_dissatisfaction'].value_counts())
print(tafe_survey_updated['dissatisfaction_with_the_department'].value_counts())

#Making a new column named "dissatisfaction" which has a value True if any of the 2 columns is True:
tafe_survey_updated["dissatisfied"] = tafe_survey_updated[['job_dissatisfaction','dissatisfaction_with_the_department']].any(axis=1, skipna=False)
tafe_survey_updated.head()
```

### 7) Combining the datasets:

Adding a column called institute to each dataset first so after combining, we know which institute the data is from:

```python
dete_survey_updated['institute'] = "DETE"
tafe_survey_updated['institute'] = "TAFE"
```

```python
#Performing Concat:

combined = pd.concat([dete_survey_updated,tafe_survey_updated], axis=0, ignore_index=True)
combined.head(20)
```

### 8) Aggregating the combined dataset based on the "institute_service" columns

Categorization based on 4 categories:<br\>
1) New: Less than 3 years at a company<br\>
2) Experienced: 3-6 years at a company<br\>
3) Established: 7-10 years at a company<br\>
4) Veteran: 11 or more years at a company

```python
def category(service):
    if service<3:
        return "New"
    elif service>=3 and service<=6:
        return "Experienced"
    elif service>= 7 and service<= 10:
        return "Established"
    else:
        return "Veteran"
    
combined['service_cat'] = combined['institute_service'].apply(category)
combined.head(20)
```

### 9) Calculating percentage of employees from the different categories who were dissatisfied:

```python
%matplotlib inline
import matplotlib.pyplot as plt

combined.loc[combined["dissatisfied"]==True,'service_cat'].value_counts(normalize=True).plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
```

### 10) Are employees who only worked for the institutes for a short period of time resigning due to some kind of dissatisfaction? 

To answer this question, we should only select rows in which the "seperation_type" column contains "Resignation" and the "dissatisfied" column = True

```python
print(combined["seperation_type"].value_counts(dropna=False))
combined_resignation = combined.loc[combined["seperation_type"].str.contains('Resignation', na=False)]
print(combined_resignation["dissatisfied"].value_counts(dropna=False))
short_employees_resignation = combined.loc[combined["service_cat"]=='New',"dissatisfied"].value_counts(normalize=True)
short_employees_resignation.plot(kind='bar')
plt.xlabel('Resignation due to Dissatisfaction')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
```

### 11) What about employees who have been there longer?

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(2,2,1)
ax1 = fig.add_subplot(2,2,2)
ax2 = fig.add_subplot(2,2,3)

experienced_employees_resignation = combined_resignation.loc[combined_resignation["service_cat"]=='Experienced',"dissatisfied"].value_counts(normalize=True)
ax = experienced_employees_resignation.plot.bar(rot=0, ax=ax)
ax.set_title("Experienced Employees")
ax.set_ylabel("Percentage")
ax.set_xlabel("Resignation due to Dissatisfaction")

established_employees_resignation = combined_resignation.loc[combined_resignation["service_cat"]=='Established', 'dissatisfied'].value_counts(normalize=True)
established_employees_resignation.plot.bar(rot=0, ax=ax1)
ax1.set_title('Established Employees')
ax1.set_ylabel('Percentage')
ax1.set_xlabel('Resignation due to Dissatisfaction')

veteran_employees_resignation = combined_resignation.loc[combined_resignation["service_cat"]=='Veteran', 'dissatisfied'].value_counts(normalize=True)
veteran_employees_resignation.plot.bar(rot=0, ax=ax2)
ax2.set_title('Established Employees')
ax2.set_ylabel('Percentage')
ax2.set_xlabel('Resignation due to Dissatisfaction')
plt.show()
```

From the data, it is apparent that a higher percetage of employees who worked for more than 3 years who resigned are dissatisfied as opposed to employees who worked for less than 3 years.
